### Correlation Matrix
This file is used to check the calculated answer in JavaScript and can also be checked via the table below if needed.

```python
import pandas as pd
columns = ['Sex', 'Length', 'Diameter', 'Height', 'Whole weight', 'Shucked weight', 'Viscera weight', 'Shell weight', 'Rings']
df = pd.read_csv('../data/abalone.data', names=columns)

df_M = df[df['Sex'] == 'M'].drop(columns=['Sex'])
df_F = df[df['Sex'] == 'F'].drop(columns=['Sex'])
df_I = df[df['Sex'] == 'I'].drop(columns=['Sex'])
```

```python
# Correlation Matrix about `Sex` equal to 'M'
df_M.corr()
```

```python
# Correlation Matrix about `Sex` equal to 'F'
df_F.corr()
```

```python
# Correlation Matrix about `Sex` equal to 'I'
df_I.corr()
```
